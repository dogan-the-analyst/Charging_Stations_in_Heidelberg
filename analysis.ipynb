```python
df.drop(columns=[
    "Vierter Ladepunkt_Public Key4",
    "Dritter Ladepunkt_Public Key3",
    "Zweiter Ladepunkt_Public Key2",
    "Erster Ladepunkt_Public Key1",
    "Adresszusatz"
], inplace=True)

df.shape
```

```python
df["Gesamt kW"] = df[[
    "Erster Ladepunkt_P1 [kW]",
    "Zweiter Ladepunkt_P2 [kW]",
    "Dritter Ladepunkt_P3 [kW]",
    "Vierter Ladepunkt_P4 [kW]",
]].sum(axis=1, skipna=True)
```

```python
df.tail()
```

```python
df['Inbetriebnahmedatum'] = pd.to_datetime('1899-12-30') + pd.to_timedelta(df['Inbetriebnahmedatum'], unit='D')
```

```python
df['Jahr'] = df['Inbetriebnahmedatum'].dt.year
```
